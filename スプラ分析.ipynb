```python
#データファイルをstatinkからDLするプログラム
import glob
import datetime
import requests
import os

def dl_csv_data():
    # ファイルを保存するフォルダを作成
    if not os.path.exists("battle-results-csv"):
        os.mkdir("battle-results-csv")

    # フォルダのパス
    path = "./battle-results-csv"

    # ファイル名のリストを取得
    filenames = glob.glob(f"{path}/*")

    # 年月日のデータを取り出す
    dates = []
    for filename in filenames:
        # ファイル名から年月日を取得
        date = filename.split("\\")[-1].split(".")[0]
        # 年月日をdatetimeオブジェクトに変換
        dt = datetime.datetime.strptime(date, "%Y-%m-%d")
        # 年月日をリストに追加
        dates.append(dt)

    # 最新の日付を取得
    if len(dates) > 0:
      latest_date = max(dates).date()
    else:
      latest_date = datetime.datetime.strptime("2022-09-25", "%Y-%m-%d").date()

    # プログラムを実行した日
    today = datetime.date.today()

    # DL対象の年月日を格納
    dl_dates = []
    # 最新の日付の翌日からプログラムを実行した日の前日までをループ
    for i in range(1, (today - latest_date).days):
        # 日付を取得
        date = latest_date + datetime.timedelta(days=i)
        # リストに追加
        dl_dates.append(date)

    for date in dl_dates:
        # DLリンクを作成
        url = f"https://dl-stats.stats.ink/splatoon-3/battle-results-csv/" + date.strftime("%Y") + "/"+ date.strftime("%m") + "/" + date.strftime("%Y-%m-%d") + ".csv"
        print(url)
        # ファイルをダウンロード
        response = requests.get(url, stream=True)
        # ファイルを保存
        with open(f"battle-results-csv/" + date.strftime("%Y-%m-%d") + ".csv", "wb") as f:
            f.write(response.content)

dl_csv_data()
```

```python
import pandas as pd
import os

def merge_csv_files():
    # フォルダ内のすべてのcsvファイルを取得
    csv_files = os.listdir("battle-results-csv")
    # csvファイルを読み込む
    dfs = []
    for csv_file in csv_files:
        df = pd.read_csv("battle-results-csv/" + csv_file)
        print(csv_file + "読み込み中")
        dfs.append(df)

    # csvファイルを統合
    df_merged = pd.concat(dfs)

    # csvファイルを書き出す
    df_merged.to_csv("battle-results.csv", index=False)

    # xmatchのcsvファイルを書きだす
    df_merged = df_merged[df_merged["lobby"]=="xmatch"]
    df_merged.to_csv("battle-results-xmatch.csv", index=False)

merge_csv_files()
```

```python
import pandas as pd
# import lightgbm as lgb
import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import numpy as np
import matplotlib.pyplot as plt
import pickle
from sklearn.model_selection import GridSearchCV

PKL_FILE_NAME = "model.pkl"

def read_csv(filename="battle-results.csv"):
    df = pd.read_csv(filename)
    df = df[["game-ver", "lobby", "mode", "stage", "power", "A1-weapon", "A2-weapon",
             "A3-weapon", "A4-weapon", "B1-weapon", "B2-weapon", "B3-weapon", "B4-weapon", "win"]]
    return df

def genModel():

    # df = read_csv("battle-results-xmatch.csv")
    df = read_csv("battle-results.csv")

    # LabelEncoderをインスタンス化する
    le = LabelEncoder()

    # カテゴリ変数を数値に変換する
    df["game-ver"] = le.fit_transform(df["game-ver"])
    df["lobby"] = le.fit_transform(df["lobby"])
    df["mode"] = le.fit_transform(df["mode"])
    df["stage"] = le.fit_transform(df["stage"])
    df["A1-weapon"] = le.fit_transform(df["A1-weapon"])
    df["A2-weapon"] = le.fit_transform(df["A2-weapon"])
    df["A3-weapon"] = le.fit_transform(df["A3-weapon"])
    df["A4-weapon"] = le.fit_transform(df["A4-weapon"])
    df["B1-weapon"] = le.fit_transform(df["B1-weapon"])
    df["B2-weapon"] = le.fit_transform(df["B2-weapon"])
    df["B3-weapon"] = le.fit_transform(df["B3-weapon"])
    df["B4-weapon"] = le.fit_transform(df["B4-weapon"])
    df["win"] = le.fit_transform(df["win"])

    print(df.head())

    X_train = df.drop(['win'], axis=1)  # X_trainはtrainのwin列以外
    Y_train = df['win']  # Y_trainはtrainのwin列

    # X_trainとY_trainをtrainとvalidに分割
    train_x, valid_x, train_y, valid_y = train_test_split(X_train, Y_train, test_size=0.33, random_state=0)

    # xgboostモデルの作成
    clf = xgb.XGBClassifier()
    # ハイパーパラメータ探索
    clf_cv = GridSearchCV(clf, {'max_depth': [2,4,6], 'n_estimators': [50,100,200]}, verbose=1)
    clf_cv.fit(train_x, train_y)
    print(clf_cv.best_params_, clf_cv.best_score_)

    # 改めて最適パラメータで学習
    clf = xgb.XGBClassifier(**clf_cv.best_params_)
    clf.fit(train_x, train_y)

    # valid_xについて推論
    oof_class = clf.predict(valid_x)  # oofはout of fold
    oof_prob = clf.predict_proba(valid_x)  # oofはout of fold
    oof = list(zip(valid_y, oof_class, oof_prob))
    oof = list(map(list, oof))

    oof.sort(key=lambda x: np.abs(x[-1][0] - 0.5), reverse=True)  # 予想の確度が高い順にソート
    print("全データテストサイズ:", len(oof))
    print('score',
          round(accuracy_score([oof[i][0] for i in range(len(oof))], [oof[i][1] for i in range(len(oof))]) * 100, 2),
          '%')  # 正解率の表示

    oof = oof[:len(oof) // 10]
    print("上位10%テストサイズ:", len(oof))
    print('score',
          round(accuracy_score([oof[i][0] for i in range(len(oof))], [oof[i][1] for i in range(len(oof))]) * 100, 2),
          '%')  # 正解率の表示

    oof = oof[:len(oof) // 10]
    print("上位1%テストサイズ:", len(oof))
    print('score',
          round(accuracy_score([oof[i][0] for i in range(len(oof))], [oof[i][1] for i in range(len(oof))]) * 100, 2),
          '%')  # 正解率の表示

    # lgb.plot_importance(gbm, max_num_features=15)
    # plt.show()


    # モデルを保存
    with open(PKL_FILE_NAME, "wb") as f:
        pickle.dump(clf, f)

def pred(data=""):
    # モデルのファイルを読み込む
    with open(PKL_FILE_NAME, "rb") as f:
        gbm = pickle.load(f)
    df = read_csv(data)
    df = df[df["lobby"]=="xmatch"]
    # print(df[df["lobby"]=="xmatch"].size)
    # LabelEncoderをインスタンス化する
    le = LabelEncoder()
    # カテゴリ変数を数値に変換する
    df["game-ver"] = le.fit_transform(df["game-ver"])
    df["lobby"] = le.fit_transform(df["lobby"])
    df["mode"] = le.fit_transform(df["mode"])
    df["stage"] = le.fit_transform(df["stage"])
    df["A1-weapon"] = le.fit_transform(df["A1-weapon"])
    df["A2-weapon"] = le.fit_transform(df["A2-weapon"])
    df["A3-weapon"] = le.fit_transform(df["A3-weapon"])
    df["A4-weapon"] = le.fit_transform(df["A4-weapon"])
    df["B1-weapon"] = le.fit_transform(df["B1-weapon"])
    df["B2-weapon"] = le.fit_transform(df["B2-weapon"])
    df["B3-weapon"] = le.fit_transform(df["B3-weapon"])
    df["B4-weapon"] = le.fit_transform(df["B4-weapon"])
    df["win"] = le.fit_transform(df["win"])
    x = df.drop(['win'], axis=1)  # X_trainはtrainのwin列以外
    y = df['win']
    clazz = gbm.predict(x, num_iteration=gbm.best_iteration_)
    # prob = gbm.predict_proba(x, num_iteration=gbm.best_iteration_)
    print('score',
          round(accuracy_score(y, clazz) * 100, 2),
          '%')

genModel()
```
